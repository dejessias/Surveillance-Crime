# Conclusion & Reflection

## Conclusion
Despite well-developed countries across the world investing in public surveillance technology to combat crime, there is little evidence that these technologies actually correlate with less crime. Even though it could be a useful tool for post-scenery crime detection, it does not tackle the structural issue of global inequalities causing crime.

As various HDI-measures seem to predict crime rates, this seems to suggest that investing in satisfactory life prospects, high quality education, and stability of income could reduce crime rates across the world. This way, surveillance cameras might even lose their policy-imposed meaning within civil and criminal discourse.
